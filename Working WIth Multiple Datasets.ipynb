```python
import numpy as np
import pandas as pd
```

```python
pop = pd.read_csv('state-population.csv')
area = pd.read_csv('state-areas.csv')
abbrev = pd.read_csv('state-abbrevs.csv')
```

```python
print(pop.head())
print(area.head())
print(abbrev.head())
```

```python
pd.merge(pop, abbrev, how = 'outer', left_on = 'state/region', right_on = 'abbreviation')
```

```python
df = pd.merge(abbrev, pop, how = 'outer', left_on = 'abbreviation', right_on = 'state/region').drop('abbreviation', axis = 1)
df
```

```python
df.isnull().any()
```

```python
df[df['population'].isnull()]
```

```python
df[df['state'].isnull()]
```

```python
#same as above code

df.loc[df['state'].isnull(), 'state/region'].unique()
```

```python
df.loc[df['state/region'] == 'PR', 'state'] = 'Puerto Rico'
df.loc[df['state/region'] == 'USA', 'state'] = 'United States'
df.isnull().any()
```

```python
df1 = pd.merge(df, area, on = 'state', how = 'left')
df1
```

```python
df1.isnull().any()
```

```python
df1['state'][df1['area (sq. mi)'].isnull()].unique()
```

```python
df1.dropna(inplace=True)
```

```python
data2010 = df1.query("year == 2010 & ages == 'total'")
data2010
```

```python
data2010.set_index('state', inplace = True)
data2010
```

```python
density = data2010['population'] / data2010['area (sq. mi)']
density.sort_values(ascending = False, inplace = True)
density
```
